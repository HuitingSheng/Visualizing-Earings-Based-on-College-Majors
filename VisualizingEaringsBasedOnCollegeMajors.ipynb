```python
# =============================================================================
# This project is for studying the relationship of College Majors and the earnings
# It is going to use Pandas,Numpy and Matplotlib.pyplot 
# It will answer the following questions:
#   Do students in more popular majors make more money?
#       Using scatter plots
#   How many majors are predominantly male? Predominantly female?
#      Using histograms
#   Which category of majors have the most students?
#      Using bar plots

# ====================================================================
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas.plotting as pplt
import numpy as np
```

```python
# read and Quick exploration of the data
recent_grads=pd.read_csv("recent-grads.csv",low_memory = 0)
print(recent_grads.shape)
pd.options.display.max_columns =25
recent_grads.head()
```

```python
recent_grads.describe()
```

```python
'''drop rows with missing values. Matplotlib expects that columns of values we pass in have matching lengths and missing values 
will cause matplotlib to throw errors.'''

recent_grads=recent_grads.dropna()
```

```python
# generate scatter plots and explore the following relations:
#   Sample_size and Median
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))

```

```python
#   Sample_size and Unemployment_rate

recent_grads.plot(x="Sample_size", y="Median", kind='scatter', title='Median vs Sample_size')
```

```python
#   Full_time and Median

recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs Sample_size')
```

```python
#   ShareWomen and Unemployment_rate
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs Full-time')
```

```python
#   Men and Median
recent_grads.plot(x='ShareWomen', y='Unemployment_rate',kind='scatter', title='Unemployment_rate vs ShareWomen')
```

```python
#   Women and Median
recent_grads.plot(x='Men', y='Median',kind='scatter', title='Median vs Men')
```

```python
#Do students in more popular majors make more money?
recent_grads.plot(x='Total', y='Median',kind='scatter', title='total vs Median')
```


students in more popular majors don't make more money

```python
#Do students that majored in subjects that were majority female make more money?
recent_grads.plot(x='ShareWomen', y='Median',kind='scatter', title='ShareWomen vs Median')
```

 students that majored in subjects that were majority female don't make more money. actually, the major have majority female make less money
 

```python
#Is there any link between the number of full-time employees and median salary?
recent_grads.plot(x='Full_time', y='Median',kind='scatter', title='Full_time vs Median')
plt.show()
```

there is no link betwwen the number of full-time employees and median salary. 

```python
#  explore the distributions of the following columns

cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,9):
    ax = fig.add_subplot(8,1,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', title=cols[r-1], rot=40)
    plt.show()
```

```python
# generate scatter_matrix plots
pplt.scatter_matrix(recent_grads[['Women', 'Men']], figsize=(6,6))

pplt.scatter_matrix(recent_grads[['Sample_size', 'Median']],  figsize=(6,6))

pplt.scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(6,6))
plt.show()
```

```python
# explor the percentages of women from the top 20 and bottom 20 majors/
recent_grads.sort_values('ShareWomen').head(20).plot.bar(x='Major', y='ShareWomen', figsize=(15,5))

recent_grads.sort_values('ShareWomen').tail(20).plot.bar(x='Major', y='ShareWomen',figsize=(15,5))
plt.show()
```

```python
# explor the percentages of unemployment_rate from the top 20 and bottom 20 majors.

recent_grads.sort_values('Unemployment_rate').head(20).plot.bar(x='Major', y='Unemployment_rate', figsize=(15,5))
plt.show()
recent_grads.sort_values('Unemployment_rate').tail(20).plot.bar(x='Major', y='Unemployment_rate', figsize=(15,5))
plt.show()
```

```python
# Use a grouped bar plot to compare the number of men with the number of women in each category of majors.
plt.figure(figsize=(15,5))
tbl = recent_grads.pivot_table(index='Major_category', values=['Men','Women'],aggfunc=np.sum)
tbl.plot.bar(x=tbl.index,y=['Men','Women'])

```

```python
# Use a box plot to explore the distributions of median salaries 
recent_grads['Median'].plot(kind='box')
```

```python
# Use a box plot to explore the distributions of unemployment rate.
recent_grads['Unemployment_rate'].plot(kind='box')
```

```python
#Use a hexagonal bin plot to visualize the columns that had dense scatter plots from earlier in the project.

recent_grads.plot(kind='hexbin', x='Full_time', y='Median')

```
